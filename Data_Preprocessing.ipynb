```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import re
```

```python
#Load the data using Pandas
df = pd.read_csv('job-market.csv')
df.head()
```

```python
#Visualize top 10 first rows
df.head(10)
```

```python
#Fix column datatypes.
df.info()
```

```python
df['Date']= pd.to_datetime(df['Date'])
df['LowestSalary'] = df['LowestSalary'].astype(float)
df['HighestSalary'] = df['HighestSalary'].astype(float)
```

```python
df.info()
```

```python
#Check and clean the data
df.isnull().sum().sum()
```

```python
#Checking for Empty Cells
count_nan_in_df = df.isnull().sum()
print (count_nan_in_df)
```

```python
#Returning Data set with no empty cells
new_df = df.dropna()
new_df
```

```python
#Finding Duplicates
duplicate = new_df.duplicated()
duplicate.value_counts()
```

```python
#Bifocating date and time
new_df['new_date'] = [d.date() for d in new_df['Date']]
new_df['new_time'] = [d.time() for d in new_df['Date']]
new_df
```

```python
#Removal of Html tags
def remove_html_tags(text):
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
new_df['new']=new_df.apply(lambda x: remove_html_tags(x['FullDescription']),axis=1)
```
